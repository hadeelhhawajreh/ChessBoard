```python

import numpy as np
import matplotlib.pyplot as plt

class ChessBoard:
    white = [1,1,1]
    blue = [0,0,1]
    red = [1,0,0]
    
    def __init__(self):
        self.grid = np.zeros([8,8,3])
        self.red_queen=None
        self.blue_queen=None
        self.attack=False
        for i in range(8):
            for j in range(8):
                if (i%2==0 and j%2==0) or (i%2==1 and j%2==1):
                    self.grid[i,j] = ChessBoard.white
        
    
          
    def render(self):
        plt.imshow(self.grid)
    
    def check_if_under_attack(self,red_x,red_y,blue_x,blue_y):
        pass
    
    def is_under_attack(self):
        # rows --> horizental
        if self.red_queen[0]==self.blue_queen[0]:
            print('is under attack ')
            self.attack=True
            return self.attack
        # cols -> vertical 
        elif  self.red_queen[1]==self.blue_queen[1]:
            print('Is Under Attack')
            self.attack=True
            return self.attack
        # diagonally
        elif np.absolute(self.red_queen[0]-self.blue_queen[0])==np.absolute(self.red_queen[1]-self.blue_queen[1]):
            print('Is Under Attack')
            self.attack=True
            return self.attack
        else:
            print('Not Under Attack')
            self.attack=False
            return self.attack
    
    def add_red(self,x,y):
        self.red_queen=(x,y)
        self.grid[x,y] = ChessBoard.red
        print(self.red_queen)
    def add_blue(self,x,y):
       self.blue_queen=(x,y)
       self.grid[x,y] = ChessBoard.blue
        

board = ChessBoard()
board.render()
```
